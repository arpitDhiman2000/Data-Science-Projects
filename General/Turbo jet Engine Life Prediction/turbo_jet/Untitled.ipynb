```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_FD004.txt", sep=' ', header=None)
df
```

```python
df = df.drop([26,27], axis='columns')
```

```python
df.columns=['ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 'SensorMeasure13', 'SensorMeasure14', 'SensorMeasure15', 'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 'SensorMeasure19', 'SensorMeasure20', 'SensorMeasure21']
```

```python
df
```

```python
df[df['ID']==100]
```

```python
EOL = []

for sl_no in df['ID']:
    EOL.append(((df[df['ID'] == sl_no]['Cycle']).values)[-1])
    
EOL
```

```python
df['EOL']=EOL
df
# EOL[len(EOL)-1]
```

```python
df['LR'] = df['Cycle'].div(df['EOL'])
df
```

```python
df = df.drop(columns=['ID','EOL'])
df
```

```python
labels = []

for i  in range(0, len(df)):
    if np.array(df['LR'])[i]<=0.6:
        labels.append(0)
    elif np.array(df['LR'])[i]<=0.8:
        labels.append(1)
    else:
        labels.append(2)
        
df['labels'] = labels

df
```

```python
df = df.drop(columns=['LR'])

df
```

```python
df.to_csv('training_004.csv', index=None)
```

```python
df1 = pd.read_csv('training_001.csv')
df2 = pd.read_csv('training_002.csv')
df3 = pd.read_csv('training_003.csv')
df4 = pd.read_csv('training_004.csv')

dff = pd.concat([df1,df2,df3,df4])

dff
```

```python
x = dff.iloc[:,0:-1]
y = dff.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split as tts

x_train, x_test, y_train, y_test = tts(x,y,test_size=0.10, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train, y_train)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_pre = RF.predict(x_test)

cm = pd.DataFrame(confusion_matrix(y_test, y_pre, normalize='true')*100)
colormap = sns.color_palette("Reds")
sns.heatmap(cm, annot=True, cmap=colormap)

accuracy_score(y_test, y_pre)
```

```python
RF.get_params()
```

```python
y_pre_f = RF.predict(x_train)

cm = pd.DataFrame(confusion_matrix(y_train, y_pre_f, normalize='true')*100)
colormap = sns.color_palette("Blues")
sns.heatmap(cm, annot=True, cmap=colormap)

accuracy_score(y_train, y_pre_f)
```
